```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/ASUS/OneDrive/Рабочий стол/Position_Salaries.csv')
```

```python
X = data.iloc[:, 1:2].values
y = data.iloc[:, 2].values
```

```python
poly_features = PolynomialFeatures(degree=3)  # Задайте бажаний ступінь полінома
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
new_positions = np.array([[6], [7]])  # Задайте нові позиції для прогнозу
new_positions_poly = poly_features.transform(new_positions)
predictions = model.predict(new_positions_poly)
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X_poly), color='red', label='Polynomial regression')
plt.scatter(new_positions, predictions, color='green', label='Predictions')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
